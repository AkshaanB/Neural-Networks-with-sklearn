```python
from sklearn import preprocessing
```

```python
label_encoding = preprocessing.LabelEncoder()
```

```python
df_spine['class'] = label_encoding.fit_transform(df_spine['class'].astype(str))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(spine,annot=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_spine.drop(['class'],axis=1)
y = df_spine['class']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_clas = MLPClassifier(hidden_layer_sizes=(100,100),
                        max_iter=100,
                        activation='relu',
                        alpha=0.0001,
                        solver='lbfgs',
                        verbose=True
                        )
```

```python
mlp_clas.fit(x_train, y_train)
```

```python
print("Training accuracy",mlp_clas.score(x_train, y_train))
```

```python
y_pred = mlp_clas.predict(x_test)
```

```python
pred_results = pd.DataFrame({'actual':y_test, 'pred':y_pred})
```

```python
pred_results.sample(10)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Testing accuracy",accuracy_score(y_test,y_pred))
```

```python
spine_data_crosstab = pd.crosstab(pred_results['pred'],pred_results['actual'])
spine_data_crosstab
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
