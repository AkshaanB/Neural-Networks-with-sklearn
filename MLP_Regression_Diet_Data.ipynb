```python
df_diet = df_diet.sample(frac=1).reset_index(drop=True)
```

```python
df_diet
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_diet.drop(['change','Date','Stone','Pounds','Ounces'],axis=1)
y = df_diet['change']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
#Standardizing numeric data to mean 0 and std 1 (unit variance)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#So, I have transformed both training and test data
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
mlp_reg = MLPRegressor(
            activation='relu',
            hidden_layer_sizes=(2,6,4),
            solver='lbfgs',
            verbose=True,
            max_iter=100
        )
```

```python
mlp_reg.fit(x_train,y_train)
```

```python
#So, the neural network is trained using back propargation to update model parameters
```

```python
y_pred = mlp_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print('Training score',mlp_reg.score(x_train,y_train))
```

```python
print('Testing score',r2_score(y_test,y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.plot(y_pred,label="Predicted",color='red')
plt.plot(y_test.values,label="Actual",color='blue')
plt.legend()
```
